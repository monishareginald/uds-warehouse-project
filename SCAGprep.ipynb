```python
import geopandas as gpd
import pandas as pd
```

```python
df = pd.read_csv('data/scag_riverside_cleaned.csv')
```

```python
df_SB = pd.read_csv('data/scag_san_bernardino_cleaned.csv')
```

```python
df_SB.head()
```

```python
#merge data frames

df = pd.concat([df,df_SB], ignore_index=True)
```

```python
df.sample(5)
```

```python
#drop geometry
df = df.drop(['PID19','CITY','LU19','LU16','JURISDICTI','SCAG_ZN_CO','Shape_Leng','Shape_Area'], axis=1)
```

```python
#get land use dummies
dummy_df = pd.get_dummies(df['LU19_CLASS'],prefix='LU19')
```

```python
dummy_df.sample(5)
```

```python
df = pd.concat([df, dummy_df], axis=1)
```

```python
df.head(5)
```

```python
df.to_csv('data/SCAG_clean.csv')
```
