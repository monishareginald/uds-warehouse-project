# SCAG Data Cleaning - Dummy Variables

```python
import pandas as pd
import geopandas as gpd
```

```python
scag_riverside_df = pd.read_csv('scag_riverside_cleaned.csv')
scag_sb_df = pd.read_csv('scag_san_bernardino_cleaned.csv')
```

```python
scag_riverside_df.head()
```

```python
scag_sb_df.head()
```

```python
scag_riverside_df['LU19_CLASS'].unique()
```

```python
scag_sb_df['LU19_CLASS'].unique()
```

```python
scag_sanbernardino_and_riverside = pd.concat([scag_riverside_df, scag_sb_df], ignore_index=True)
```

```python
scag_sanbernardino_and_riverside.head()
```

```python
scag_sanbernardino_and_riverside['LU19_CLASS'].unique()
```

```python
lu19_class_dummy = pd.get_dummies(scag_sanbernardino_and_riverside['LU19_CLASS'])
```

```python
lu19_class_dummy.head()
```

```python
scag_dummy = scag_sanbernardino_and_riverside.join(lu19_class_dummy)
```

```python
scag_dummy.sample(5)
```

```python
scag_dummy.to_csv('scag_dummy.csv', index=False)
```
