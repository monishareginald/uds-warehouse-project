```python
scag_dummy = pd.read_csv('scag_dummy.csv')
```

```python
# Takes about 7 minutes to run
parcels = gpd.read_file('parcels_warehouses.gpkg')
```

```python
parcels.head()
```

```python
# scag_dummy[scag_dummy.APN19 == 'RIVER']

scag_dummy = scag_dummy[scag_dummy['APN19'] != 'RIVER']
```

```python
scag_dummy = scag_dummy.astype({"APN19": float})
```

```python
scag_dummy.info()
```

```python
parcels.info()
```

### Left join

```python
join_scag_to_parcels_left = pd.merge(parcels, scag_dummy, left_on='APN', right_on='APN19', how='left')
```

### Calculate $ per square acre

```python
join_scag_to_parcels_left_2019['ACRES']
```

```python
join_scag_to_parcels_left_2019['LAND_VALUE']
```

```python
dollars_per_acre = join_scag_to_parcels_left_2019['LAND_VALUE']/join_scag_to_parcels_left_2019['ACRES']
```

```python
dollars_per_acre
```

```python
join_scag_to_parcels_left_2019['dollars_per_acre'] = dollars_per_acre
```

```python
join_scag_to_parcels_left_2019.head()
```

```python
join_scag_to_parcels_left_2019 = join_scag_to_parcels_left_2019.drop(columns=['geometry_y'])
```

```python
join_scag_to_parcels_left_2019.rename(columns={'geometry_x':'geometry'}, inplace=True)
```

```python
join_scag_to_parcels_left_2019.info()
```

```python
join_scag_to_parcels_left_2019['geometry']
```

```python
join_scag_to_parcels_left_2019.to_csv('join_scag_to_parcels_left_2019.csv', index=False)
```
